```python
import razor.flow as rf
from razor.api import datasources
import numpy as np
import logging
logger = logging.getLogger()
```

```python
@rf.block
class DefaultVal:
    a: int = 1
        
    def run(self):
        logger.info(self.a)
```

```python
config = {'cores': 1, 'ram': 4096}

def_val = DefaultVal()
def_val.executor = rf.ContainerExecutor(cores=config['cores'], memory=config['ram'])

pipeline = rf.Pipeline(name='default_value', targets=[def_val])
pipeline.show()
```
