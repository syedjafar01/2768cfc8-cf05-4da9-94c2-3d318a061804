```python
import razor.flow as rf
from razor.api import datasources
import numpy as np
import logging
logger = logging.getLogger()
```

```python
@rf.block
class DefaultType:
    a: float
        
    def run(self):
        logger.info(self.a)
```

```python
config = {'cores': 1, 'ram': 4096}

def_type = DefaultType(a=7.0)
def_type.executor = rf.ContainerExecutor(cores=config['cores'], memory=config['ram'])

pipeline = rf.Pipeline(name='default_float', targets=[def_type])
pipeline.show()
```

```python
def create_pipeline():
    return pipeline
```

```python
razor.api.engines('consistencyEngine').execute(pipeline)
```
