```python
import razor.flow as rf
import time
from typing import List, Dict, Generic, TypeVar, Optional, Any, Type, Iterable, TypeVar

@rf.block
class Add:
    input1: int    
    input2: int
    result: rf.Output [int] = rf.Output(label='Result', transport=rf.KafkaTransport)
    def run(self):
        self.logger.setLevel(10)
        self.logger.info(f'input1  : {self.input1}')
        self.logger.info(f'input2 : {self.input2}')
        self.logger.debug('debug mesage')
        self.logger.info(f'self.logger.level : {self.logger.level}')
        self.logger.info(f'self.logger._level : {dir(self.logger)}')
        self.logger.info(f'self.logger.getEffectiveLevel : {self.logger.getEffectiveLevel()}')
        self.logger.info(f'self.logger.setLevel : {dir(self.logger.setLevel)}')
        self.logger.warning('warning mesage')
        self.logger.info('info mesage')
        self.logger.error('error mesage')
#         time.sleep(60)
        print('print statement')
        nsum=self.input1 + self.input2
        self.result.put(nsum)
```

```python
import logging
logging.DEBUG
```

```python
a=Add(input1=1,input2=2)
```

```python
p=rf.Pipeline(targets=[a])
```
